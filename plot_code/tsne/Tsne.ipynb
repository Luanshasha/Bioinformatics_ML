```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
sns.set()###
plt.rc('font',family='Times New Roman') ###Times New Roman
plt.figure(figsize=(12,25), dpi=300,facecolor=(1, 1, 1))#dpi300,dpi300

#adjust distrances of subplots
plt.subplots_adjust(wspace =0.3, hspace =0.4)# 
model=TSNE(n_components=2,random_state=0)

ax2=plt.subplot(621)
# data1=pd.read_csv(r"C:\Users\Administrator\Desktop\Peptides_optim2\Anti-cancer_Peptides\fastadata\Tse\Bit20(NT=2).csv",header=None,index_col=None)
# labels1=pd.read_csv(r"C:\Users\Administrator\Desktop\Peptides_optim2\Anti-cancer_Peptides\fastadata\Tse\Bit20(NT=2)2.csv",header=None,index_col=None)
data1=pd.read_csv(r"./kmer_2_kmer_3_binary.csv",header=None,index_col=None)
labels1=pd.read_csv(r"./first_label.csv",header=None,index_col=None)
tsne_data1=model.fit_transform(data1)
tsne_data1=np.vstack((tsne_data1.T,labels1.T)).T
tsne_df=pd.DataFrame(data=tsne_data1,columns=("Dimension1","Dimension2","label"))
p1=tsne_df[(tsne_df.label=="Positive") ]
p2=tsne_df[(tsne_df.label=="Negative") ]
x1=p1.values[:,0]
y1=p1.values[:,1]
x2=p2.values[:,0]
y2=p2.values[:,1]
##plt.plot(x, y1)
plt.plot(x1, y1,'o', color='#EE7621',label = 'Positive',markersize='4')
#2
plt.plot(x2, y2,'o', color='#36648B',label = 'Negative',markersize='4')
# ax.scatter(x1,y1,c = 'r',marker = 'o')  
# ax.scatter(x2,y2,c = 'b',marker = 'o')  
plt.xlabel('Dimension1',fontsize=9)
plt.ylabel('Dimension2',fontsize=9)
plt.title('hm5C_before',fontsize=12)
plt.legend(loc="upper right")
plt.text(-79,67,'(A)',fontsize='13')
plt.xlim((-60, 60))###before_tsne.ipynbDimension1
plt.ylim((-60, 75))###before_tsne.ipynbDimension2


ax2=plt.subplot(623)
# data1=pd.read_csv(r"C:\Users\Administrator\Desktop\Peptides_optim2\Anti-cancer_Peptides\fastadata\Tse\Bit20(NT=2).csv",header=None,index_col=None)
# labels1=pd.read_csv(r"C:\Users\Administrator\Desktop\Peptides_optim2\Anti-cancer_Peptides\fastadata\Tse\Bit20(NT=2)2.csv",header=None,index_col=None)
data1=pd.read_csv(r"./26.csv",header=None,index_col=None)
labels1=pd.read_csv(r"./first_label.csv",header=None,index_col=None)
tsne_data1=model.fit_transform(data1)
tsne_data1=np.vstack((tsne_data1.T,labels1.T)).T
tsne_df=pd.DataFrame(data=tsne_data1,columns=("Dimension1","Dimension2","label"))
p1=tsne_df[(tsne_df.label=="Positive") ]
p2=tsne_df[(tsne_df.label=="Negative") ]
x1=p1.values[:,0]
y1=p1.values[:,1]
x2=p2.values[:,0]
y2=p2.values[:,1]
##plt.plot(x, y1)
plt.plot(x1, y1,'o', color='#EE7621',label = 'Positive',markersize='4')
#2
plt.plot(x2, y2,'o', color='#36648B',label = 'Negative',markersize='4')
# ax.scatter(x1,y1,c = 'r',marker = 'o')  
# ax.scatter(x2,y2,c = 'b',marker = 'o')  
plt.xlabel('Dimension1',fontsize=9)
plt.ylabel('Dimension2',fontsize=9)
plt.title('hm5C_after',fontsize=12)
plt.legend(loc="upper right")
plt.text(-79,57,'(B)',fontsize='13')
plt.xlim((-60, 60))
plt.ylim((-70, 65))






#plt.show()
plt.savefig('./tsne4.pdf',bbox_inches='tight')#plt.showsvg,pdf,tiff


```
