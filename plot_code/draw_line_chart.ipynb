```python
# used as plot program
import matplotlib.pyplot as plt
import numpy as np
import math
plt.figure(figsize=(15,6), dpi=300,facecolor=(1, 1, 1))#figsize
# plt.tight_layout()#
plt.subplots_adjust(wspace =0.3, hspace =0.5)#


plt.subplot(231)

x=np.arange(1,10,0.01)
# plt.plot(x,np.sin(x))
file_location="./data/draw_line_chart/S1.txt"
all_results=[]
with open(file_location,'r') as f:#r
    data=f.readlines()
    for yy in data:
        if yy.startswith("[[["):
            all_results.append(float(yy.strip()[3:-3].split(",")[1]))#
all_results
#x=range(len(all_results))
x= list(map(lambda x: x*10, range(len(all_results))))
length_x=len(x)
max_value=max(all_results)
max_value_index=all_results.index(max_value)
max_value=round(max_value,4)

#plt.annotataxytext
plt.annotate("["+str(max_value_index*10+1)+","+str(max_value)+"]",xytext=(max_value_index*10-0.2*length_x,max_value),xy=(max_value_index*10,max_value),fontsize=11)
plt.plot([max_value_index*10,max_value_index*10],[0,max_value],color="red",label="Best ACC: %.4f"%max_value)#

plt.plot(x,all_results)#
plt.title("C.elegans",fontsize=12)#
plt.legend(loc="upper right",fontsize=11)#legend

#

plt.xlim([0, length_x*10]) #x
plt.ylim((0.5, 1))#  y
#
plt.ylabel('Accuracy',fontsize=11) #y
plt.xlabel('Feature dimension',fontsize=11)#x
#
my_y_ticks = np.arange(0.5, 1, 0.1)
plt.yticks(my_y_ticks) #y
my_x_ticks = np.arange(0, length_x*10, 100)
plt.xticks(my_x_ticks)#x
plt.text(-0.19*10*length_x*1.0,0.98,'(A)',fontsize='xx-large') #


plt.subplot(232)

x=np.arange(1,10,0.01)
# plt.plot(x,np.sin(x))
file_location="./data/draw_line_chart/S2.txt"
all_results=[]
with open(file_location,'r') as f:#r
    data=f.readlines()
    for yy in data:
        if yy.startswith("[[["):
            all_results.append(float(yy.strip()[3:-3].split(",")[1]))#
all_results
#x=range(len(all_results))
x= list(map(lambda x: x*10, range(len(all_results))))
length_x=len(x)
max_value=max(all_results)
max_value_index=all_results.index(max_value)
max_value=round(max_value,4)

#plt.annotataxytext
plt.annotate("["+str(max_value_index*10+1)+","+str(max_value)+"]",xytext=(max_value_index*10-0.2*length_x,max_value),xy=(max_value_index*10,max_value),fontsize=11)
plt.plot([max_value_index*10,max_value_index*10],[0,max_value],color="red",label="Best ACC: %.4f"%max_value)#

plt.plot(x,all_results)#
plt.title("D.melanogaster",fontsize=12)#
plt.legend(loc="upper right",fontsize=11)#legend

#

plt.xlim([0, length_x*10]) #x
plt.ylim((0.5, 1))#  y
#
plt.ylabel('Accuracy',fontsize=11) #y
plt.xlabel('Feature dimension',fontsize=11)#x
#
my_y_ticks = np.arange(0.5, 1, 0.1)
plt.yticks(my_y_ticks) #y
my_x_ticks = np.arange(0, length_x*10, 100)
plt.xticks(my_x_ticks)#x
plt.text(-0.19*10*length_x*1.0,0.98,'(B)',fontsize='xx-large') #

plt.subplot(233)

x=np.arange(1,10,0.01)
# plt.plot(x,np.sin(x))
file_location="./data/draw_line_chart/S3.txt"
all_results=[]
with open(file_location,'r') as f:#r
    data=f.readlines()
    for yy in data:
        if yy.startswith("[[["):
            all_results.append(float(yy.strip()[3:-3].split(",")[1]))#
all_results
#x=range(len(all_results))
x= list(map(lambda x: x*10, range(len(all_results))))
length_x=len(x)
max_value=max(all_results)
max_value_index=all_results.index(max_value)
max_value=round(max_value,4)

#plt.annotataxytext
plt.annotate("["+str(max_value_index*10+1)+","+str(max_value)+"]",xytext=(max_value_index*10-0.2*length_x,max_value),xy=(max_value_index*10,max_value),fontsize=11)
plt.plot([max_value_index*10,max_value_index*10],[0,max_value],color="red",label="Best ACC: %.4f"%max_value)#

plt.plot(x,all_results)#
plt.title("A.thaliana",fontsize=12)#
plt.legend(loc="upper right",fontsize=11)#legend

#

plt.xlim([0, length_x*10]) #x
plt.ylim((0.5, 1))#  y
#
plt.ylabel('Accuracy',fontsize=11) #y
plt.xlabel('Feature dimension',fontsize=11)#x
#
my_y_ticks = np.arange(0.5, 1, 0.1)
plt.yticks(my_y_ticks) #y
my_x_ticks = np.arange(0, length_x*10, 100)
plt.xticks(my_x_ticks)#x
plt.text(-0.19*10*length_x*1.0,0.98,'(C)',fontsize='xx-large') #


plt.subplot(234)

x=np.arange(1,10,0.01)
# plt.plot(x,np.sin(x))
file_location="./data/draw_line_chart/S4.txt"
all_results=[]
with open(file_location,'r') as f:#r
    data=f.readlines()
    for yy in data:
        if yy.startswith("[[["):
            all_results.append(float(yy.strip()[3:-3].split(",")[1]))#
all_results
#x=range(len(all_results))
x= list(map(lambda x: x*10, range(len(all_results))))
length_x=len(x)
max_value=max(all_results)
max_value_index=all_results.index(max_value)
max_value=round(max_value,4)

#plt.annotataxytext
plt.annotate("["+str(max_value_index*10+1)+","+str(max_value)+"]",xytext=(max_value_index*10-0.2*length_x,max_value),xy=(max_value_index*10,max_value),fontsize=11)
plt.plot([max_value_index*10,max_value_index*10],[0,max_value],color="red",label="Best ACC: %.4f"%max_value)#

plt.plot(x,all_results)#
plt.title("E.coli",fontsize=12)#
plt.legend(loc="upper right",fontsize=11)#legend

#

plt.xlim([0, length_x*10]) #x
plt.ylim((0.5, 1))#  y
#
plt.ylabel('Accuracy',fontsize=11) #y
plt.xlabel('Feature dimension',fontsize=11)#x
#
my_y_ticks = np.arange(0.5, 1, 0.1)
plt.yticks(my_y_ticks) #y
my_x_ticks = np.arange(0, length_x*10, 100)
plt.xticks(my_x_ticks)#x
plt.text(-0.19*10*length_x*1.0,0.98,'(D)',fontsize='xx-large') #


plt.subplot(235)

x=np.arange(1,10,0.01)
# plt.plot(x,np.sin(x))
file_location="./data/draw_line_chart/S5.txt"
all_results=[]
with open(file_location,'r') as f:#r
    data=f.readlines()
    for yy in data:
        if yy.startswith("[[["):
            all_results.append(float(yy.strip()[3:-3].split(",")[1]))#
all_results
#x=range(len(all_results))
x= list(map(lambda x: x*10, range(len(all_results))))
length_x=len(x)
max_value=max(all_results)
max_value_index=all_results.index(max_value)
max_value=round(max_value,4)

#plt.annotataxytext
plt.annotate("["+str(max_value_index*10+1)+","+str(max_value)+"]",xytext=(max_value_index*10-0.2*length_x,max_value),xy=(max_value_index*10,max_value),fontsize=11)
plt.plot([max_value_index*10,max_value_index*10],[0,max_value],color="red",label="Best ACC: %.4f"%max_value)#

plt.plot(x,all_results)#
plt.title("G.subterraneus",fontsize=12)#
plt.legend(loc="upper right",fontsize=11)#legend

#

plt.xlim([0, length_x*10]) #x
plt.ylim((0.5, 1))#  y
#
plt.ylabel('Accuracy',fontsize=11) #y
plt.xlabel('Feature dimension',fontsize=11)#x
#
my_y_ticks = np.arange(0.5, 1, 0.1)
plt.yticks(my_y_ticks) #y
my_x_ticks = np.arange(0, length_x*10, 100)
plt.xticks(my_x_ticks)#x
plt.text(-0.19*10*length_x*1.0,0.98,'(E)',fontsize='xx-large') #


plt.subplot(236)

x=np.arange(1,10,0.01)
# plt.plot(x,np.sin(x))
file_location="./data/draw_line_chart/S6.txt"
all_results=[]
with open(file_location,'r') as f:#r
    data=f.readlines()
    for yy in data:
        if yy.startswith("[[["):
            all_results.append(float(yy.strip()[3:-3].split(",")[1]))#
all_results
#x=range(len(all_results))
x= list(map(lambda x: x*10, range(len(all_results))))
length_x=len(x)
max_value=max(all_results)
max_value_index=all_results.index(max_value)
max_value=round(max_value,4)

#plt.annotataxytext
plt.annotate("["+str(max_value_index*10+1)+","+str(max_value)+"]",xytext=(max_value_index*10-0.2*length_x,max_value),xy=(max_value_index*10,max_value),fontsize=11)
plt.plot([max_value_index*10,max_value_index*10],[0,max_value],color="red",label="Best ACC: %.4f"%max_value)#

plt.plot(x,all_results)#
plt.title("G.pickeringii",fontsize=12)#
plt.legend(loc="upper right",fontsize=11)#legend

#

plt.xlim([0, length_x*10]) #x
plt.ylim((0.5, 1))#  y
#
plt.ylabel('Accuracy',fontsize=11) #y
plt.xlabel('Feature dimension',fontsize=11)#x
#
my_y_ticks = np.arange(0.5, 1, 0.1)
plt.yticks(my_y_ticks) #y
my_x_ticks = np.arange(0, length_x*10, 100)
plt.xticks(my_x_ticks)#x
plt.text(-0.19*10*length_x*1.0,0.98,'(F)',fontsize='xx-large') #




#plt.savefig('./BestACC_line_chart.pdf',bbox_inches='tight') # pdf, svg,tiff bbox_inches
 #print "done"
plt.show()#plt.savefig
```
